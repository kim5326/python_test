```python
import pandas as pd
```

```python
train = pd.read_csv('../data/titanic/train.csv')

train.shape
```

```python
train.head()
```

```python
# 시각화 -> seaborn

%matplotlib inline

# 세부적인 시각화 가능 

import seaborn as sns
```

# countplot

```python
sns.countplot(data= train, x = 'Pclass')
```

```python
sns.countplot(data= train, x = 'Pclass', hue = 'Survived')
```

```python
sns.countplot(data= train, x = 'Embarked', hue = 'Survived')
```

# Barplot

```python
sns.barplot(data = train, x = 'Pclass', y = 'Fare')
```

```python
sns.barplot(data = train, x = 'Pclass', y = 'Fare', hue = 'Survived')


# Pclass 1 이면서 죽은 사람들의 Fare는 검은색 바 범위 안에 있다 라고 추정 가능
```

# Pointplot

```python
sns.pointplot(data = train, x = 'Pclass', y = 'Fare', hue = 'Survived')

# 연속적인 변화를 볼때 추세적인 변화 확인하는데 유용
```

## lmplot : 산점도

```python
sns.lmplot(data = train, x = 'Age', y = 'Fare', hue = 'Survived')
```

## displot

```python
sns.distplot(train['Fare'])
```
